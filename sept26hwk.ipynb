````python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# Load the dataset
penguin_data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Drop rows with missing values
penguin_data = penguin_data.dropna(subset=['flipper_length_mm', 'species'])

# Define a function to get statistical information and add lines/rectangles
def add_statistics_to_fig(fig, species_df, species_name):
    # Calculate statistical measures
    mean_val = species_df['flipper_length_mm'].mean()
    median_val = species_df['flipper_length_mm'].median()
    min_val = species_df['flipper_length_mm'].min()
    max_val = species_df['flipper_length_mm'].max()
    std_val = species_df['flipper_length_mm'].std()
    q1 = species_df['flipper_length_mm'].quantile(0.25)
    q3 = species_df['flipper_length_mm'].quantile(0.75)

    # Add vertical lines for mean and median
    fig.add_vline(x=mean_val, line_dash="dash", line_color="blue", annotation_text=f"{species_name} Mean", annotation_position="top left")
    fig.add_vline(x=median_val, line_dash="dot", line_color="green", annotation_text=f"{species_name} Median", annotation_position="top left")

    # Add rectangles for range, IQR, and 2 standard deviations
    # Range
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor="lightgray", opacity=0.2, line_width=0, annotation_text=f"{species_name} Range", annotation_position="bottom right")

    # IQR
    fig.add_vrect(x0=q1, x1=q3, fillcolor="yellow", opacity=0.3, line_width=0, annotation_text=f"{species_name} IQR", annotation_position="bottom right")

    # 2 standard deviations from the mean
    fig.add_vrect(x0=mean_val - 2*std_val, x1=mean_val + 2*std_val, fillcolor="red", opacity=0.2, line_width=0, annotation_text=f"{species_name} ±2 Std", annotation_position="bottom right")

# Create a histogram for each species
species_list = penguin_data['species'].unique()
fig = go.Figure()

for species in species_list:
    species_df = penguin_data[penguin_data['species'] == species]
    fig.add_trace(go.Histogram(x=species_df['flipper_length_mm'], name=species, opacity=0.75))

    # Add statistical markers to the figure
    add_statistics_to_fig(fig, species_df, species)

# Update layout
fig.update_layout(barmode='overlay', title="Flipper Length Distribution with Statistical Markers", xaxis_title="Flipper Length (mm)", yaxis_title="Count")

# Show the figure
fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the penguins dataset
penguin_data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
penguin_data = penguin_data.dropna(subset=['flipper_length_mm', 'species'])

# Set up the figure and axes for a row of three KDE plots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

# Define a function to add statistical markers to the KDE plot
def add_statistics(ax, species_df):
    # Calculate statistical measures
    mean_val = species_df['flipper_length_mm'].mean()
    median_val = species_df['flipper_length_mm'].median()
    min_val = species_df['flipper_length_mm'].min()
    max_val = species_df['flipper_length_mm'].max()
    std_val = species_df['flipper_length_mm'].std()
    q1 = species_df['flipper_length_mm'].quantile(0.25)
    q3 = species_df['flipper_length_mm'].quantile(0.75)

    # Add vertical lines for mean and median
    ax.axvline(mean_val, color='blue', linestyle='--', label=f'Mean: {mean_val:.1f}')
    ax.axvline(median_val, color='green', linestyle='-.', label=f'Median: {median_val:.1f}')
    
    # Add shaded areas for range, IQR, and 2 standard deviations
    # Range (min to max)
    ax.axvspan(min_val, max_val, color='gray', alpha=0.2, label=f'Range: [{min_val:.1f}, {max_val:.1f}]')

    # IQR (Q1 to Q3)
    ax.axvspan(q1, q3, color='yellow', alpha=0.3, label=f'IQR: [{q1:.1f}, {q3:.1f}]')

    # 2 standard deviations from the mean
    ax.axvspan(mean_val - 2 * std_val, mean_val + 2 * std_val, color='red', alpha=0.2, label=f'±2 Std Dev')

# Loop over the species and plot the KDEs
species_list = penguin_data['species'].unique()

for i, species in enumerate(species_list):
    species_df = penguin_data[penguin_data['species'] == species]
    ax = axes[i]

    # Plot KDE using seaborn
    sns.kdeplot(species_df['flipper_length_mm'], ax=ax, fill=True, color='skyblue', label=f'{species}', bw_adjust=0.6)
    
    # Add statistical markers
    add_statistics(ax, species_df)

    # Set plot labels and title
    ax.set_title(f'Flipper Length Distribution - {species}')
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Density')
    ax.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````

````python
# Generating right-skewed (gamma) data
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Right-Skewed Distribution")
# Showing the histogram for right-skewed data
fig1.show(renderer="png")

# Calculating the mean and median for the right-skewed sample
mean_sample1 = sample1.mean()
median_sample1 = np.quantile(sample1, [0.5])[0]

# Generating left-skewed (negative gamma) data
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Left-Skewed Distribution")
# Showing the histogram for left-skewed data
fig2.show(renderer="png")

# Calculating the mean and median for the left-skewed sample
mean_sample2 = sample2.mean()
median_sample2 = np.quantile(sample2, [0.5])[0]

# Printing the mean and median for both skewed datasets
print("Right-skewed distribution (Sample 1): Mean =", mean_sample1, ", Median =", median_sample1)
print("Left-skewed distribution (Sample 2): Mean =", mean_sample2, ", Median =", median_sample2)
````

Mean and Median:

    Mean: The average of the dataset, sensitive to extreme values.
    Median: The middle value when the data is ordered, less affected by extreme values.

Skewness: Refers to the asymmetry in the distribution of data.

Left-Skewed (Negative Skewness): The tail on the left side of the distribution is longer or fatter than the right side. In a left-skewed distribution, the mean is typically less than the median because the lower values pull the mean downward.
Order: Mean < Median < Mode


Right-Skewed (Positive Skewness): The tail on the right side of the distribution is longer or fatter than the left side. In a right-skewed distribution, the mean is greater than the median because the higher values pull the mean upward. Order: Mode < Median < Mean

````python
import pandas as pd
import plotly.express as px

# Load the baby names dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Make identical boy and girl names distinct
bn['name'] = bn['name'] + " " + bn['sex']  

# Rank names by year based on prevalence
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)

# Sort values
bn = bn.sort_values(['name', 'year'])

# Calculate percent change in name prevalence
bn['percent change'] = bn['percent'].diff()

# Identify new names for resetting percent change
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent'] 

# Sort by year
bn = bn.sort_values('year')

# Restrict to "common" names
bn = bn[bn.percent > 0.001]

# Filter to only show positive percent changes
bn = bn[bn['percent change'] > 0]  # Only keep positive percent change values

# Create the scatter plot
fig = px.scatter(bn, 
                 x="percent change", 
                 y="rank", 
                 animation_frame="year", 
                 animation_group="name", 
                 size="percent", 
                 color="sex", 
                 hover_name="name", 
                 size_max=50, 
                 range_x=[0, bn['percent change'].max()])  # Adjust range_x to start at 0

# Update the y-axis to reverse the order for rank
fig.update_yaxes(autorange='reversed')

# Show the figure for submissions
fig.show(renderer="png")  # Use `fig.show(renderer="png")` for GitHub and MarkUs submissions
````
